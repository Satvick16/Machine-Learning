```
# TODO add a legend with the order of each line added to the figure

import matplotlib.pyplot as plt
import numpy as np
from sympy import *
import time


def update(m, slope_sign, prev_x0):

    x = Symbol('x')
    f = x ** 2 - 4

    f_prime = f.diff(x)

    f_prime = lambdify(x, f_prime)
    f = lambdify(x, f)

    if slope_sign == 0:
        exit()
    elif slope_sign < 0:
        x0 = prev_x0 + 0.25 * (abs(m))
    else:
        x0 = prev_x0 - 0.25 * (abs(m))

    try:
        y0 = f(x0)
        m = f_prime(x0)

        if m > 0:
            slope_sign = 1
        elif m < 0:
            slope_sign = -1
        else:
            slope_sign = 0

        ########################################################################

        x = np.linspace(-4, 4, 999)
        y = m * x - m * x0 + y0

        plt.plot(x, y)

        # recursive call with updated x0
        update(m, slope_sign, x0)

    except UnboundLocalError:
        exit()
    except RecursionError:
        exit()

################################################################################

x = np.linspace(-4, 4, 999)

# hypothetical loss function
f = x ** 2 - 4

colormap = plt.cm.Set1
plt.gca().set_prop_cycle(plt.cycler('color', plt.cm.jet(np.linspace(0, 1, 15))))

plt.xlim(-4, 4)
plt.ylim(-5, 5)

plt.plot(x, f, c='black')

initial_x0 = -2.5
x = Symbol('x')
f = x ** 2 - 4
f_prime = f.diff(x)
f_prime = lambdify(x, f_prime)
m = f_prime(initial_x0)

if m > 0:
    initial_slope_sign = 1
elif m < 0:
    initial_slope_sign = -1
else:
    initial_slope_sign = 0

update(m, initial_slope_sign, initial_x0)
```
